```python
import pickle

# 指定文件路径
file_path = 'llama3.3fold2_train_epoch1_18k1.pkl'

# 打开文件并加载数据
with open(file_path, 'rb') as file:
    data1 = pickle.load(file)
```

```python
file2_path = 'llama3.3fold3_train_epoch1_18k1.pkl'

# 打开文件并加载数据
with open(file2_path, 'rb') as file:
    data2 = pickle.load(file)
```

```python
import numpy as np
data1['winner'] = np.where(data1['winner_model_a'] > data1['winner_model_b'], 'model_a', 'model_b')
data2['winner'] = np.where(data2['winner_model_a'] > data2['winner_model_b'], 'model_a', 'model_b')
```

```python
data1
```

```python
data2
```

```python
data1 = data1[['id', 'prompt', 'response_a', 'response_b', 'model_a', 'model_b', 'language', 'winner']]
data2 = data2[['id', 'prompt', 'response_a', 'response_b', 'model_a', 'model_b', 'language', 'winner']]
```

```python
import pandas as pd
merged_data = pd.merge(data1, data2, on=['id', 'prompt', 'response_a', 'response_b', 'model_a', 'model_b', 'language', 'winner'], indicator=True)
```

```python
same_rows = merged_data[merged_data['_merge'] == 'both']
```

```python
same_rows
```

```python
final_data2 = same_rows[['id', 'prompt', 'response_a', 'response_b', 'model_a', 'model_b', 'language', 'winner']]
```
